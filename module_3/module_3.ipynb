```python
import numpy as np
import pandas as pd
from scipy import stats
np.random.seed(42)

n_samples = 50
group_1 = np.random.normal(loc=5.0, scale=2.0, size=n_samples)
group_2 = np.random.normal(loc=6.5, scale=2.0, size=n_samples) # Independent shift
group_1_post = group_1 + np.random.normal(loc=1.5, scale=1.0, size=n_samples) # Paired shift

# Student's t-test
t_stat_ind, p_val_t_ind = stats.ttest_ind(group_1, group_2)
print(f"Independent t-test:       t = {t_stat_ind:.4f}, p = {p_val_t_ind:.4f}")

# Mann-Whitney U test
u_stat, p_val_mw = stats.mannwhitneyu(group_1, group_2)
print(f"Mann-Whitney U test:      U = {u_stat:.4f}, p = {p_val_mw:.4f}")

# Paired Student's t-test
t_stat_rel, p_val_t_rel = stats.ttest_rel(group_1, group_1_post)
print(f"Paired t-test:            t = {t_stat_rel:.4f}, p = {p_val_t_rel:.4f}")

# Wilcoxon signed-rank test
w_stat, p_val_wilc = stats.wilcoxon(group_1, group_1_post)
print(f"Wilcoxon signed-rank:     W = {w_stat:.4f}, p = {p_val_wilc:.4f}")
```
